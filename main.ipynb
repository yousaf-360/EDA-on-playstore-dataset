```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv('googleplaystore.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.dropna(inplace=True)
df
```

```python
plt.figure(figsize=(9,5))
plt.xlabel('Category')
plt.xticks(rotation=90)
plt.ylabel('Installs')
plt.bar(df['Category'],df['Installs'])
plt.show()
```

```python
plt.xlabel('Rating')
plt.ylabel('Installs')
plt.scatter(df['Rating'],df['Installs'])
plt.show()
```

```python
#category with most apps
counter=df['Category'].value_counts()
top5=counter.head(5)
plt.bar(top5.index, top5.values)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Top 5 Categories')
plt.show()
```

```python
#relation between rating and reviwes
plt.ylabel('rating')
plt.xlabel('reviews')
plt.xticks(rotation=90)
plt.scatter(df['Reviews'],df['Rating'])
plt.show()
```

```python
#relation between rating and category
plt.ylabel('rating')
plt.xlabel('Category')
plt.xticks(rotation=90)
plt.bar(df['Category'],df['Rating'])
plt.show()
```

```python
#categories sorted based on their ratings
grouped=df.groupby('Category')['Rating'].mean()
grouped = grouped.sort_values(ascending=False)
grouped
```
